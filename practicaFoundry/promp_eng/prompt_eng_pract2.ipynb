### 2.1 - Experimentación con temperature
Prueba el **mismo prompt** con diferentes valores de `temperature`:
- `temperature = 0.0`
- `temperature = 0.5`
- `temperature = 1.0`
- `temperature = 1.5` (si la API lo permite)

Analiza cómo cambian las respuestas. Usa un caso práctico (ejemplo: generar un eslogan, escribir código, responder una pregunta técnica).

```python
import os

from openai import OpenAI
import json

# Configuración
api_key = os.getenv("Key")
base_url = os.getenv("base_url")
client = OpenAI(base_url=base_url, api_key=api_key)

print("="*80)
print("🎛️  SECCIÓN 2.1: EXPERIMENTACIÓN CON TEMPERATURE")
print("="*80)

print("\n📝 OBJETIVO:")
print("Entender cómo temperature afecta la creatividad y consistencia de las respuestas")

print("\n" + "-"*80)
print("- CASO PRÁCTICO: Generar Eslóganes para una Startup de IA")
print("-"*80)

prompt_base = """Genera 3 eslóganes creativos y memorables para una startup de IA llamada "NexusAI".

Requisitos:
- Máximo 10 palabras cada uno
- Deben ser llamativos y únicos
- Transmitir innovación y confiabilidad

Formato:
1. [eslogan]
2. [eslogan]
3. [eslogan]
"""

print("\n📋 PROMPT BASE:")
print("-" * 80)
print(prompt_base)

# Temperaturas a probar
temperaturas = [0.0, 0.5, 1.0, 1.5]

resultados_temperatura = {}

print("\n" + "="*80)
print("🔬 EXPERIMENTACIÓN CON 4 VALORES DE TEMPERATURE")
print("="*80)

for temp in temperaturas:
    print(f"\n{'━'*80}")
    print(f"🌡️  TEMPERATURE = {temp}")
    print('━'*80)
    
    print(f"\n📊 DESCRIPCIÓN ESPERADA:")
    if temp == 0.0:
        desc = "Determinístico: Siempre MISMA respuesta, sin variación"
    elif temp == 0.5:
        desc = "Conservador: Poco variado, respuestas similares, predecible"
    elif temp == 1.0:
        desc = "Balanceado: Buena mezcla de creatividad y coherencia (DEFAULT)"
    else:  # 1.5
        desc = "Creativo: Muy variado, pueden haber respuestas extrañas"
    
    print(f"   {desc}")
    
    # Generar respuestas MÚLTIPLES para ver la variación
    print(f"\n🔄 GENERANDO 3 INTENTOS para ver consistencia...\n")
    
    respuestas_intento = []
    
    for intento in range(3):
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt_base}],
            temperature=temp,
            max_tokens=150,
            top_p=1.0  # Mantener constante
        )
        
        respuesta = response.choices[0].message.content
        respuestas_intento.append(respuesta)
        
        print(f"INTENTO {intento + 1}:")
        print("-" * 40)
        print(respuesta)
        print()
    
    resultados_temperatura[temp] = respuestas_intento

print("\n" + "="*80)
print("📊 ANÁLISIS COMPARATIVO: VARIABILIDAD Y CONSISTENCIA")
print("="*80)

print(f"""
┌──────────────────────────────────────────────────────────────────┐
│ COMPARACIÓN DE RESULTADOS BY TEMPERATURE                         │
└──────────────────────────────────────────────────────────────────┘

1️⃣  TEMPERATURE = 0.0 (DETERMINÍSTICO):
   ├─ Predicibilidad:    ✅ PERFECTA (siempre igual)
   ├─ Variación:         ❌ CERO (todas idénticas)
   ├─ Creatividad:       ❌ NULA (respuesta fija)
   ├─ Riesgo:            ❌ Aburrido, poco inspirador
   ├─ Caso de uso:       ✅ Extracción de datos, QA, clasificación
   └─ Observación:       Las 3 respuestas son IDÉNTICAS

2️⃣  TEMPERATURE = 0.5 (CONSERVADOR):
   ├─ Predicibilidad:    ✅ MUY ALTA (mostly similar)
   ├─ Variación:         ⚠️  BAJA (pequeñas diferencias)
   ├─ Creatividad:       ⚠️  MEDIA (variación controlada)
   ├─ Riesgo:            ⚠️  Predecible pero poco sorprendente
   ├─ Caso de uso:       ✅ Resúmenes, síntesis, traducciones
   └─ Observación:       Las respuestas son similares, pocos cambios

3️⃣  TEMPERATURE = 1.0 (BALANCEADO):
   ├─ Predicibilidad:    ⚠️  BUENA (coherente pero variable)
   ├─ Variación:         ✅ MEDIA (cambios interesantes)
   ├─ Creatividad:       ✅ BUENA (sin perder coherencia)
   ├─ Riesgo:            ✅ Bajo (respuestas sensatas)
   ├─ Caso de uso:       ✅ DEFAULT para casi todo
   │                     ✅ Ideación, brainstorming
   │                     ✅ Contenido creativo
   └─ Observación:       Las respuestas son DISTINTAS pero relevantes

4️⃣  TEMPERATURE = 1.5 (CREATIVO/EXPERIMENTAL):
   ├─ Predicibilidad:    ❌ BAJA (muy variado)
   ├─ Variación:         ✅ ALTA (respuestas muy diferentes)
   ├─ Creatividad:       ✅✅ MÁXIMA (puede ser poco convencional)
   ├─ Riesgo:            ⚠️  MEDIO-ALTO (puede no tener sentido)
   ├─ Caso de uso:       ✅ Creatividad extrema
   │                     ✅ Lluvia de ideas radical
   │                     ❌ NO para tareas críticas
   └─ Observación:       Las respuestas pueden ser incoherentes

┌──────────────────────────────────────────────────────────────────┐
│ CONCLUSIÓN VISUAL: Temperature controla CREATIVIDAD vs COHERENCIA│
└──────────────────────────────────────────────────────────────────┘

ESPECTRO:
0.0 ◄──────────────────────────────────────────────► 1.5+
 ↓                                                    ↓
DETERMINÍSTICO                        CAÓTICO/EXPERIMENTAL
SIN VARIACIÓN                         MÁS CREATIVIDAD
PERFECTO PARA DATOS                   PARA BRAINSTORMING

┌──────────────────────────────────────────────────────────────────┐
│ RECOMENDACIONES POR CASO DE USO:                                 │
└──────────────────────────────────────────────────────────────────┘

✅ temperature = 0.0-0.2:
   • Extracción de datos estructurados
   • Clasificación (necesita consistencia 100%)
   • Análisis de sentimientos
   • Traducciones precisas
   • Respuestas a preguntas factales

✅ temperature = 0.5-0.8:
   • Resúmenes y síntesis
   • Paraphrasing
   • Generación de código
   • Documentación técnica
   • Respuestas a preguntas complejas

✅ temperature = 1.0 (DEFAULT):
   • Email y comunicación profesional
   • Artículos y blog posts
   • Content marketing
   • Explicaciones educativas
   • Ideación general

✅ temperature = 1.2-1.5:
   • Brainstorming y lluvia de ideas
   • Escritura creativa (historias, poesía)
   • Marketing copy experimental
   • Names y slogans innovadores
   • Generación de problemas para tests

💡 REGLA DE ORO:
   Mayor temperatura = Mayor riesgo pero mayor originalidad
   Menor temperatura = Mayor confiabilidad pero menor sorpresa
   
   Ajusta temperature según TOLERANCIA AL RIESGO de tu aplicación.
""")

print("\n" + "="*80)
print("🔍 MI REFLEXIÓN COMO ESTUDIANTE:")
print("="*80)
print("""
DESCUBRIMIENTOS PERSONALES:

1. Temperature es MÁS IMPORTANTE que otros parámetros
   → Controla el "comportamiento" del modelo entero
   → Pequeños cambios tienen BIG IMPACT

2. No existe un "mejor valor"
   → Depende TOTALMENTE de tu caso de uso
   → Experimenta y ajusta según necesidad

3. Temperatura 0.0 NO es siempre lo mejor
   → Parece bueno (consistencia), PERO es robótico
   → Para naturalidad, necesitas al menos 0.7+

4. Temperature 1.5 es "potencialmente arriesgada"
   → Puede generar respuestas que no importan
   → Pero es EXCELENTE para lluvia de ideas

5. El DEFAULT (1.0) es sorprendentemente versátil
   → OpenAI lo eligió por razón: funciona bien
   → A menos que tengas necesidad específica, déjalo

CONSEJO PRO:
Empieza con 1.0. Baja si necesitas consistencia.
Sube si necesitas más creatividad.
Nunca sobre-optimices: la diferencia 0.8 vs 0.9 es mínima.
""")
```

### 2.2 - Experimentación con top_p
Prueba el **mismo prompt** con diferentes valores de `top_p`:
- `top_p = 0.1`
- `top_p = 0.5`
- `top_p = 0.9`
- `top_p = 1.0`

Mantén `temperature = 1.0` para ver el efecto puro de top_p. Compara con los resultados de temperature.

```python
import os

from openai import OpenAI

# Reutilizar configuración anterior
api_key = os.getenv("Key")
base_url = os.getenv("base_url")
client = OpenAI(base_url=base_url, api_key=api_key)

print("\n" + "="*80)
print("🎛️  SECCIÓN 2.2: EXPERIMENTACIÓN CON TOP_P")
print("="*80)

print("\n📝 OBJETIVO:")
print("Entender cómo top_p (nucleus sampling) afecta la diversidad de palabras elegidas")

print("\n" + "-"*80)
print("- CASO PRÁCTICO: Describir un Producto de Forma Natural")
print("-"*80)

prompt_base_top_p = """Describe brevemente un smartphone moderno de premium en 2-3 oraciones.
Haz que suene natural y atractivo para un cliente.

Tu descripción:"""

print("\n📋 PROMPT BASE:")
print("-" * 80)
print(prompt_base_top_p)

# Top_p valores a probar (mantener temperature = 1.0 para ver efecto puro)
top_p_valores = [0.1, 0.5, 0.9, 1.0]

resultados_top_p = {}

print("\n" + "="*80)
print("🔬 EXPERIMENTACIÓN CON 4 VALORES DE TOP_P (temperature = 1.0)")
print("="*80)

for top_p in top_p_valores:
    print(f"\n{'━'*80}")
    print(f"📊 TOP_P = {top_p}")
    print('━'*80)
    
    # Explicación de cada valor
    if top_p == 0.1:
        desc = "EXTREMADAMENTE RESTRICTIVO: Solo el 10% más probable"
    elif top_p == 0.5:
        desc = "RESTRICTIVO: Solo el 50% más probable"
    elif top_p == 0.9:
        desc = "PERMISIVO: El 90% más probable"
    else:  # 1.0
        desc = "COMPLETO: Todos los tokens posibles (sin filtro)"
    
    print(f"\n📌 INTERPRETACIÓN:")
    print(f"   {desc}")
    print(f"   → Resultado esperado: Variedad controlada")
    
    # Generar 3 intentos para ver variación
    print(f"\n🔄 GENERANDO 3 INTENTOS...\n")
    
    respuestas_intento_top_p = []
    
    for intento in range(3):
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt_base_top_p}],
            temperature=1.0,  # FIJO
            top_p=top_p,      # VARIABLE
            max_tokens=100
        )
        
        respuesta = response.choices[0].message.content
        respuestas_intento_top_p.append(respuesta)
        
        print(f"INTENTO {intento + 1}:")
        print("-" * 40)
        print(respuesta)
        print()
    
    resultados_top_p[top_p] = respuestas_intento_top_p

print("\n" + "="*80)
print("📊 ANÁLISIS COMPARATIVO: TOP_P vs TEMPERATURA")
print("="*80)

print(f"""
┌──────────────────────────────────────────────────────────────────┐
│ COMPARACIÓN DE RESULTADOS BY TOP_P (Temperature = 1.0 FIJO)      │
└──────────────────────────────────────────────────────────────────┘

1️⃣  TOP_P = 0.1 (ULTRA-RESTRICTIVO):
   ├─ Palabras seleccionadas: ❌ SOLO las más obvias (top 10%)
   ├─ Diversidad:             ❌ MÍNIMA (respuestas casi idénticas)
   ├─ Predicción:             ✅ EXCELENTE (sabes qué esperar)
   ├─ Naturalidad:            ❌ MENOS NATURAL (palabras repetidas)
   ├─ Riesgo:                 ⚠️  Aburrido, predecible
   ├─ Utilidad:               ⚠️  Solo para máxima consistencia
   └─ Observación:            Las 3 respuestas son CASI IDÉNTICAS

2️⃣  TOP_P = 0.5 (RESTRICTIVO):
   ├─ Palabras seleccionadas: ⚠️  50% más probable
   ├─ Diversidad:             ⚠️  BAJA-MEDIA (alguna variación)
   ├─ Predicción:             ✅ BUENA (bastante predecible)
   ├─ Naturalidad:            ⚠️  Algo forzada, palabras obvias
   ├─ Riesgo:                 ⚠️  Puede sonar robótica
   ├─ Utilidad:               ✅ Cuando necesitas consistencia
   │                          ✅ Documentación técnica
   └─ Observación:            Variación pequeña, controlada

3️⃣  TOP_P = 0.9 (PERMISIVO):
   ├─ Palabras seleccionadas: ✅ 90% más probable
   ├─ Diversidad:             ✅ BUENA (variación natural)
   ├─ Predicción:             ✅ ACEPTABLE (coherente)
   ├─ Naturalidad:            ✅ MÁS NATURAL (variedad sin riesgo)
   ├─ Riesgo:                 ✅ BAJO (respuestas sensatas)
   ├─ Utilidad:               ✅ EXCELENTE para textos naturales
   │                          ✅ Recomendado para la mayoría
   └─ Observación:            Las respuestas VARÍAN pero coherentes

4️⃣  TOP_P = 1.0 (SIN FILTRO):
   ├─ Palabras seleccionadas: ✅ TODAS las posibles
   ├─ Diversidad:             ✅ MÁXIMA (máxima variación)
   ├─ Predicción:             ❌ IMPREDECIBLE
   ├─ Naturalidad:            ✅ Natural pero a veces extraño
   ├─ Riesgo:                 ⚠️  MEDIO (puede salirse de guión)
   ├─ Utilidad:               ✅ Creatividad, brainstorming
   │                          ❌ NO para tareas estructuradas
   └─ Observación:            Las respuestas pueden ser MUY diferentes

┌──────────────────────────────────────────────────────────────────┐
│ TOP_P vs TEMPERATURE: ¿CUÁL ES LA DIFERENCIA?                    │
└──────────────────────────────────────────────────────────────────┘

TEMPERATURA:
  • Controla CÓMO DE LOCO es el modelo
  • Afecta la probabilidad de TODOS los tokens
  • Valor alto = modelo más osado/arriesgado
  • Baja = modelo más cauteloso

TOP_P:
  • Controla CUÁLES tokens están disponibles
  • Filtra palabras menos probables
  • Reduce "ruido" lingüístico
  • No afecta "audacia" del modelo

ANALOGÍA:
  Temperature = Volumen de la música
  Top_p = Tipo de música disponible

┌──────────────────────────────────────────────────────────────────┐
│ ¿POR QUÉ NO AJUSTAR AMBOS AL MISMO TIEMPO?                       │
└──────────────────────────────────────────────────────────────────┘

❌ PROBLEMA:
   • Temperature alta + top_p bajo = CONTRADICCIÓN
   • Le dices: "Sé creativo" (temp) + "Pero solo palabras obvias" (top_p)
   • Resultado: Comportamiento impredecible y difícil de controlar

✅ SOLUCIÓN:
   • Si quieres más creatividad: ↑ Temperature
   • Si quieres más control: ↓ Temperature
   • Si quieres diversidad natural: ↓ Top_p (0.8-0.9)
   • Si quieres consistencia: ↓ Top_p

📌 RECOMENDACIÓN OFICIAL (OpenAI):
   NO ajustar ambos simultáneamente.
   Elige UNA estrategia:
   
   Opción A: Usar TEMPERATURE para control
             (Mantener top_p = 1.0)
   
   Opción B: Usar TOP_P para control
             (Mantener temperature = 1.0)
   
   Opción A es más común y intuitiva.

┌──────────────────────────────────────────────────────────────────┐
│ RECOMENDACIONES POR CASO DE USO:                                 │
└──────────────────────────────────────────────────────────────────┘

✅ top_p = 0.1-0.3:
   • Extracción de información
   • Clasificación binaria
   • QA (cuando la respuesta es clara)
   • Cuando necesitas máxima consistencia

✅ top_p = 0.5-0.7:
   • Generación de código
   • Resúmenes técnicos
   • Análisis de sentimientos
   • Cuando quieres control + algo de naturalidad

✅ top_p = 0.8-0.95:
   • Texto natural (descripciones, artículos)
   • Email y comunicación
   • Content marketing
   • La mayoría de tareas creativas
   • RECOMENDADO por defecto

✅ top_p = 1.0:
   • Máxima libertad creativa
   • Research y brainstorming
   • Cuando el "ruido" es bienvenido
   • Raro usarlo (o usa temperature en su lugar)

💡 REGLA DE ORO:
   Si no sabes qué usar: top_p = 0.9 (casi universal)
   La mayoría de modelos funcionan mejor con 0.85-0.95
""")

print("\n" + "="*80)
print("🔍 MI REFLEXIÓN COMO ESTUDIANTE:")
print("="*80)
print("""
DESCUBRIMIENTOS SOBRE TOP_P:

1. Top_p es MENOS INTUITIVO que temperature
   → Temperature es "creatividad" (fácil de entender)
   → Top_p es "filtro de palabras" (más técnico)

2. Top_p NO reemplaza temperatura
   → Sirven propósitos DIFERENTES
   → No deberías ajustar ambos
   → Elige uno como estrategia principal

3. El efecto de top_p es MÁS SUTIL
   → Cambios de 0.8 a 0.9 = casi ninguna diferencia
   → Solo valores extremos (0.1 vs 1.0) son obvios
   → Por eso temperature es más común en práctica

4. Top_p es ÚTIL para casos específicos
   → Cuando necesitas naturalidad sin creatividad
   → Cuando quieres evitar "palabras raras"
   → En dominios técnicos donde palabras importan

5. OpenAI recomienda: Usa TEMPERATURE primero
   → Top_p es más para ajustes finos
   → A menos que tengas caso específico

CONSEJO PRO:
Empieza con temperature.
Solo toca top_p si necesitas:
  ✓ Máxima consistencia (bájalo a 0.7)
  ✓ Naturalidad pura (súbelo a 0.95)
La mayoría del tiempo: temperature = 1.0, top_p = 0.9
""")
```

### 2.3 - Experimentación con Penalties
Prueba prompts que **tiendan a repetir contenido** (por ejemplo: describir un producto, generar múltiples ideas similares) con:
- `presence_penalty = 0.0` vs `presence_penalty = 0.6`
- `frequency_penalty = 0.0` vs `frequency_penalty = 0.8`
- Combinación de ambos penalties

Analiza qué tipo de repeticiones evita cada uno.

```python
import os
from openai import OpenAI

# Reutilizar configuración
api_key = os.getenv("Key")
base_url = os.getenv("base_url")
client = OpenAI(base_url=base_url, api_key=api_key)

print("\n" + "="*80)
print("🎛️  SECCIÓN 2.3: EXPERIMENTACIÓN CON PENALTIES")
print("="*80)

print("\n📝 OBJETIVO:")
print("Entender presence_penalty y frequency_penalty para evitar repeticiones")

print("\n" + "-"*80)
print("- CASO PRÁCTICO: Describir Características de una Laptop (Repetición Controlada)")
print("-"*80)

# Este prompt tiende a repetir según cómo responda
prompt_penalties = """Describe las 5 características principales de una laptop gaming moderna.
Sé descriptivo y detallado en cada característica."""

print("\n📋 PROMPT BASE (diseñado para provocar repeticiones):")
print("-" * 80)
print(prompt_penalties)

print("\n" + "="*80)
print("🔬 PARTE 1: PRESENCE_PENALTY")
print("="*80)

print("""
📌 QUÉ HACE:
   • Penaliza palabras que YA APARECIERON en la respuesta
   • NO importa CUÁNTAS veces aparecieron
   • Una penalización = "-1 punto" por token anterior
   • Rango: -2.0 a +2.0 (0 = sin penalización)

CASOS:
   presence_penalty = 0.0 → Las palabras pueden repetirse sin costo
   presence_penalty = 0.6 → Penaliza un poco si la palabra ya apareció
   presence_penalty = 2.0 → Penaliza MUCHO si la palabra ya apareció
""")

presence_configs = [
    (0.0, "SIN PENALTY (permite repetición)"),
    (0.6, "MEDIUM PENALTY (desalienta repetición)"),
]

print("\n" + "🔄 GENERANDO RESPUESTAS CON DIFERENTES PRESENCE_PENALTY...\n")

resultados_presence = {}

for presence_pen, descripcion in presence_configs:
    print(f"\n{'━'*80}")
    print(f"🎯 PRESENCE_PENALTY = {presence_pen} ({descripcion})")
    print('━'*80)
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt_penalties}],
        temperature=1.0,
        top_p=0.9,
        presence_penalty=presence_pen,
        frequency_penalty=0.0,  # Mantener constante
        max_tokens=200
    )
    
    respuesta = response.choices[0].message.content
    resultados_presence[presence_pen] = respuesta
    
    print(f"\n💬 RESPUESTA:")
    print("-" * 40)
    print(respuesta)
    
    # Análisis simple de repeticiones
    palabras = respuesta.lower().split()
    palabras_unicas = len(set(palabras))
    palabras_totales = len(palabras)
    repeticion_ratio = (1 - (palabras_unicas / palabras_totales)) * 100
    
    print(f"\n📊 MÉTRICAS:")
    print(f"   • Palabras totales: {palabras_totales}")
    print(f"   • Palabras únicas: {palabras_unicas}")
    print(f"   • Ratio repetición: {repeticion_ratio:.1f}%")

print("\n" + "="*80)
print("🔬 PARTE 2: FREQUENCY_PENALTY")
print("="*80)

print("""
📌 QUÉ HACE:
   • Penaliza palabras según CUÁNTAS VECES aparecieron
   • Cuantas más repeticiones, mayor penalización
   • La penalización es PROPORCIONAL a frecuencia
   • Rango: -2.0 a +2.0 (0 = sin penalización)

CASOS:
   frequency_penalty = 0.0 → Las palabras pueden repetirse sin límite
   frequency_penalty = 0.6 → Penaliza más cuanto más se repite
   frequency_penalty = 0.8 → Penaliza FUERTEMENTE las repeticiones
""")

frequency_configs = [
    (0.0, "SIN PENALTY (permite repetición frecuente)"),
    (0.8, "HIGH PENALTY (evita mucho la repetición)"),
]

print("\n" + "🔄 GENERANDO RESPUESTAS CON DIFERENTES FREQUENCY_PENALTY...\n")

resultados_frequency = {}

for freq_pen, descripcion in frequency_configs:
    print(f"\n{'━'*80}")
    print(f"🎯 FREQUENCY_PENALTY = {freq_pen} ({descripcion})")
    print('━'*80)
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt_penalties}],
        temperature=1.0,
        top_p=0.9,
        presence_penalty=0.0,  # Mantener constante
        frequency_penalty=freq_pen,  # VARIABLE
        max_tokens=200
    )
    
    respuesta = response.choices[0].message.content
    resultados_frequency[freq_pen] = respuesta
    
    print(f"\n💬 RESPUESTA:")
    print("-" * 40)
    print(respuesta)
    
    # Análisis simple de repeticiones
    palabras = respuesta.lower().split()
    palabras_unicas = len(set(palabras))
    palabras_totales = len(palabras)
    repeticion_ratio = (1 - (palabras_unicas / palabras_totales)) * 100
    
    print(f"\n📊 MÉTRICAS:")
    print(f"   • Palabras totales: {palabras_totales}")
    print(f"   • Palabras únicas: {palabras_unicas}")
    print(f"   • Ratio repetición: {repeticion_ratio:.1f}%")

print("\n" + "="*80)
print("🔬 PARTE 3: COMBINANDO AMBOS PENALTIES")
print("="*80)

print("\n🔄 GENERANDO RESPUESTA CON AMBOS PENALTIES ALTOS...\n")

print(f"\n{'━'*80}")
print(f"🎯 presence_penalty = 0.6 + frequency_penalty = 0.8")
print('━'*80)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt_penalties}],
    temperature=1.0,
    top_p=0.9,
    presence_penalty=0.6,
    frequency_penalty=0.8,
    max_tokens=200
)

respuesta_combined = response.choices[0].message.content

print(f"\n💬 RESPUESTA:")
print("-" * 40)
print(respuesta_combined)

palabras = respuesta_combined.lower().split()
palabras_unicas = len(set(palabras))
palabras_totales = len(palabras)
repeticion_ratio = (1 - (palabras_unicas / palabras_totales)) * 100

print(f"\n📊 MÉTRICAS:")
print(f"   • Palabras totales: {palabras_totales}")
print(f"   • Palabras únicas: {palabras_unicas}")
print(f"   • Ratio repetición: {repeticion_ratio:.1f}%")

print("\n" + "="*80)
print("📊 ANÁLISIS COMPARATIVO: PENALTIES")
print("="*80)

print(f"""
┌──────────────────────────────────────────────────────────────────┐
│ COMPARACIÓN: PRESENCE vs FREQUENCY PENALTY                       │
└──────────────────────────────────────────────────────────────────┘

PRESENCE_PENALTY (Penaliza SI/NO la palabra apareció):
┌─────────────────────────────────────────┐
│ presence = 0.0 vs presence = 0.6         │
├─────────────────────────────────────────┤
│ 0.0:                                    │
│   ✓ Palabras pueden repetirse           │
│   ✗ Respuesta más natural pero repetida │
│   ✗ Menos variedad de vocabulario       │
│   • Bueno para: Fluidez, naturalidad    │
│                                         │
│ 0.6:                                    │
│   ✓ Evita repetir CUALQUIER palabra     │
│   ✓ Vocabulario más variado             │
│   ✗ Menos natural (suena forzado)       │
│   • Bueno para: Variedad, originalidad  │
└─────────────────────────────────────────┘

FREQUENCY_PENALTY (Penaliza CUÁNTO se repitió):
┌─────────────────────────────────────────┐
│ frequency = 0.0 vs frequency = 0.8      │
├─────────────────────────────────────────┤
│ 0.0:                                    │
│   ✓ Sin restricción                     │
│   ✗ Palabras pueden repetirse mucho     │
│   ✗ Puede sonar monótono                │
│   • Bueno para: Énfasis, importancia    │
│                                         │
│ 0.8:                                    │
│   ✓ Evita repeticiones múltiples        │
│   ✓ Más variado, less monotonous        │
│   ✗ Puede sacrificar naturalidad        │
│   • Bueno para: Documentación variada   │
└─────────────────────────────────────────┘

┌──────────────────────────────────────────────────────────────────┐
│ ¿CUÁL ES LA DIFERENCIA CLAVE?                                    │
└──────────────────────────────────────────────────────────────────┘

Ejemplo con palabra "importante":

Sin penalties (0.0, 0.0):
"Es importante... importante... importante notar que es importante..."
→ Repetida 4 veces, suena terrible

Presence_penalty = 0.6:
"Es importante... Debe notar que es crucial... Es fundamental..."
→ Primera vez OK, luego evita CUALQUIER repetición
→ Usa sinónimos

Frequency_penalty = 0.8:
"Es importante... importante pero... aunque importante es también..."
→ Penalización proporcional: 1° vez OK, 2° vez penalties, 3° vez MÁS penalties
→ Gradualmente desalienta

Ambos penalties (0.6, 0.8):
"Es importante... Deberíamos considerar... Esto es fundamental..."
→ Máxima variedad, NO repeticiones

┌──────────────────────────────────────────────────────────────────┐
│ RECOMENDACIONES POR CASO DE USO:                                 │
└──────────────────────────────────────────────────────────────────┘

✅ BAJA O CERO PENALTIES (0.0-0.2):
   • Código (puede necesitar repetir variables)
   • Tutoriales (repetir conceptos es útil)
   • Documentación técnica (términos específicos)
   • Cuando la repetición tiene propósito

✅ MEDIUM PRESENCE, BAJO FREQUENCY (0.3-0.5, 0.0):
   • Resúmenes profesionales
   • News articles
   • Análisis técnicos

✅ MEDIUM PENALTIES (0.5, 0.5):
   • Blog posts
   • Email marketing
   • Descripciones de productos
   • Content creativo general

✅ ALTOS PENALTIES (0.6+, 0.6+):
   • Nombre de caracteres de historias
   • Generación de ideas únicas
   • Brainstorming creativo
   • Cuando lo "plano" es enemigo

💡 REGLA DE ORO:
   presence_penalty = evita repetir CUALQUIER palabra
   frequency_penalty = permite algo pero penaliza MUCHO
   
   Para mayoría de casos: ambos en 0.5-0.6
   Para máxima variedad: ambos en 0.8+
   NUNCA ambos en 0.0 (aburrido)
""")

print("\n" + "="*80)
print("🔍 MI REFLEXIÓN COMO ESTUDIANTE:")
print("="*80)
print("""
DESCUBRIMIENTOS SOBRE PENALTIES:

1. Los penalties son MENOS UTILIZADOS que temperature/top_p
   → Pero son INCREÍBLEMENTE ÚTILES cuando los necesitas
   → La mayoría de la gente no sabe que existen

2. Presence_penalty y frequency_penalty sirven propósitos DIFERENTES
   → No son lo mismo, aunque ambos evitan repetición
   → Presencia = "¿apareció?" vs Frecuencia = "¿cuántas veces?"

3. Los penalties SACRIFICAN un poco naturalidad por variedad
   → Si subes presencia_penalty, la respuesta suena más forzada
   → Es balance entre naturalidad y originalidad

4. Penalties son PERFECTOS para casos específicos
   → Generación de nombres (presence_penalty alto)
   → Brainstorming (ambos altos)
   → Documentación (frequency_penalty moderado)

5. Combinación de penalties es PODEROSA
   → No necesitas usarlos separados
   → O ambos a 0, o ambos elevados
   → No mezcles (0.0 presencia + 0.8 frecuencia sería raro)

CONSEJO PRO:
Los penalties son "especializados".
Úsalos cuando:
  • Necesitas evitar repeticiones
  • Quieres variedad forzada
  • Generación de ideas/nombres
  
No los uses cuando:
  • El contenido necesita precisión (código)
  • Repetición tiene propósito (énfasis)
  • Quieres máxima naturalidad
""")
```
