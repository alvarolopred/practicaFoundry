### 1️⃣ **Rol / Persona** 🎭

**📝 Explicación (Personal):** Es como si estuvieras diciendo la IA  que tenga una personalidad especifica para las respuestas

````python
import os

from openai import OpenAI

# Configuración del modelo
api_key = os.getenv("Key")
base_url = os.getenv("base_url")  # Asegúrate de configurar esta variable de entorno con tu URL personalizada
client = OpenAI(base_url=base_url, api_key=api_key)

print("="*80)
print("🎭 TÉCNICA: ROL / PERSONA")
print("="*80)

print("\n🔬 EJEMPLO PRÁCTICO:")
print("-" * 80)

# Pregunta base
pregunta = "¿Cuál es la importancia del machine learning en la actualidad?"

# Diferentes roles para la misma pregunta
roles = [
    ("SIN ROL (baseline)", None),
    ("📊 Analista de datos", "Eres un analista de datos experto con 10 años de experiencia."),
    ("👨‍🏫 Profesor universitario", "Eres un profesor universitario de Ciencias de Datos explicando a estudiantes principiantes."),
    ("💼 Ejecutivo de startup", "Eres un CEO de una startup de IA buscando justificar inversión en ML."),
    ("🤖 Ingeniero de IA", "Eres un ingeniero de IA experto en sistemas de deep learning y optimización."),
    ("🔬 Investigador/Académico en IA", "Eres un investigador académico de renombre internacional en Inteligencia Artificial publicando en conferencias top-tier."),
]

for rol_nombre, sistema_prompt in roles:
    print(f"\n{'='*80}")
    print(f"🎯 {rol_nombre}")
    print('='*80)
    
    if sistema_prompt:
        print(f"📋 SISTEMA: {sistema_prompt}")
    else:
        print("📋 SISTEMA: (ninguno - respuesta genérica)")
    
    print(f"\n❓ PREGUNTA: {pregunta}")
    
    # Construir el mensaje
    if sistema_prompt:
        messages = [
            {"role": "system", "content": sistema_prompt},
            {"role": "user", "content": pregunta}
        ]
    else:
        messages = [
            {"role": "user", "content": pregunta}
        ]
    
    # Llamar al modelo
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        max_tokens=300,
        temperature=0.7
    )
    
    print(f"\n💬 RESPUESTA:")
    print("-" * 80)
    print(response.choices[0].message.content)

print("\n" + "="*80)
print("📊 ANÁLISIS COMPARATIVO:")
print("="*80)
print("""
✅ OBSERVACIONES CLAVE:

1️⃣ SIN ROL (baseline):
   → Respuesta genérica, equilibrada, accesible pero poco específica.
   → No hay perspectiva clara ni enfoque determinado.

2️⃣ ANALISTA DE DATOS:
   → Enfoque en métricas, KPIs y toma de decisiones basada en datos.
   → Lenguaje práctico y orientado a negocios.

3️⃣ PROFESOR UNIVERSITARIO:
   → Explicación accesible con analogías y ejemplos educativos.
   → Diseñado para que entiendan principiantes.

4️⃣ CEO DE STARTUP:
   → Énfasis en ROI, ventajas competitivas y oportunidades de negocio.
   → Lenguaje de inversión y viabilidad comercial.

5️⃣ INGENIERO DE IA:
   → Detalles técnicos, arquitecturas, algoritmos y optimización.
   → Comprensión profunda de los componentes internos.

6️⃣ INVESTIGADOR/ACADÉMICO EN IA:
   → Perspectiva teórica y de investigación punta.
   → Énfasis en innovación, limitaciones frontales, y direcciones futuras.
   → Referencia a trabajos recientes y tendencias en la comunidad académica.

🎯 DIFERENCIA CLAVE (CON ROL vs SIN ROL):
→ Sin rol: Respuesta plana, sin carácter.
→ Con rol: Respuesta personalizada, con perspectiva clara y lenguaje específico.

El modelo ADAPTA:
✓ Profundidad técnica
✓ Tono y registro del lenguaje
✓ Ejemplos y referencias
✓ Énfasis en diferentes aspectos
✓ Audiencia objetivo implícita

💡 CONCLUSIÓN:
La técnica Role/Persona es EXTREMADAMENTE PODEROSA para:
- Generar contenido personalizado sin reescribir el prompt base
- Adaptar nivel de tecnicidad automáticamente
- Cambiar perspectiva y enfoque de la respuesta
- Mejorar relevancia para públicos específicos
- Crear variaciones de contenido de forma eficiente
""")
````

### 2️⃣ **Especificidad (be explicit)** 🎯

**📝 Explicación (Personal):** Ser concreto, es decir claras y concisas.

````python
import os

from openai import OpenAI

api_key = os.getenv("Key")
base_url = os.getenv("base_url")
client = OpenAI(base_url=base_url, api_key=api_key)

print("="*80)
print("🎯 TÉCNICA: ESPECIFICIDAD (Be Explicit)")
print("="*80)

print("\n🔬 EJEMPLO PRÁCTICO:")
print("-" * 80)

prompts = [
    ("POCO ESPECÍFICO", "Dame información sobre inteligencia artificial"),
    
    ("MEDIO ESPECÍFICO", 
     "Dame un párrafo sobre inteligencia artificial y su impacto en medicina (máximo 150 palabras)."),
    
    ("MUY ESPECÍFICO",
     """Dame una respuesta CON ESTA ESTRUCTURA EXACTA:
     
TÍTULO: Inteligencia Artificial en Medicina

SECCIONES:
1. Definición (máx 40 palabras)
2. Tres aplicaciones principales (bullet points)
3. Ventajas (máx 60 palabras)
4. Desafíos actuales (máx 50 palabras)
5. Conclusión (máx 30 palabras)

REQUISITOS:
- Lenguaje técnico pero accesible
- Incluir al menos 2 ejemplos reales
- Total máximo 300 palabras"""),
]

respuestas_spec = []

for nivel, prompt in prompts:
    print(f"\n{'='*80}")
    print(f"📋 {nivel.upper()}")
    print('='*80)
    print(f"\n🔤 PROMPT:\n{prompt}")
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=400,
        temperature=0.7
    )
    
    respuesta = response.choices[0].message.content
    respuestas_spec.append(respuesta)
    
    print(f"\n💬 RESPUESTA:")
    print("-" * 80)
    print(respuesta)
    print(f"\n📊 MÉTRICAS: {len(respuesta)} caracteres | {respuesta.count(chr(10)) + 1} líneas")

print("\n" + "="*80)
print("📊 ANÁLISIS COMPARATIVO VISUAL:")
print("="*80)

print(f"""
┌─────────────────────────────────────────────────────────────────────┐
│ COMPARACIÓN DE ESPECIFICIDAD                                        │
└─────────────────────────────────────────────────────────────────────┘

1️⃣ POCO ESPECÍFICO:
   ├─ Estructura:       ❌ SIN ESTRUCTURA
   ├─ Predicción:       ❌ IMPREDECIBLE
   ├─ Longitud:         {len(respuestas_spec[0]):>3} caracteres
   ├─ Formato:          ❌ FLEXIBLE (cualquier forma es válida)
   └─ Automación:       ❌ MUY DIFÍCIL de procesar

2️⃣ MEDIO ESPECÍFICO:
   ├─ Estructura:       ⚠️  PARCIALMENTE DEFINIDA
   ├─ Predicción:       ⚠️  ALGO PREDECIBLE
   ├─ Longitud:         {len(respuestas_spec[1]):>3} caracteres  ({((len(respuestas_spec[1]) - len(respuestas_spec[0])) / len(respuestas_spec[0]) * 100):.0f}% más)
   ├─ Formato:          ⚠️  SEMI-FLEXIBLE (puede variar)
   └─ Automación:       ⚠️  POSIBLE pero frágil

3️⃣ MUY ESPECÍFICO:
   ├─ Estructura:       ✅ 100% COMO PEDISTE
   ├─ Predicción:       ✅ GARANTIZADO
   ├─ Longitud:         {len(respuestas_spec[2]):>3} caracteres  ({((len(respuestas_spec[2]) - len(respuestas_spec[0])) / len(respuestas_spec[0]) * 100):.0f}% más)
   ├─ Formato:          ✅ EXACTO Y CONSISTENTE
   └─ Automación:       ✅ PERFECTO para procesar

┌─────────────────────────────────────────────────────────────────────┐
│ CLAVE: Mayor especificidad = Mayor CONTROL y PREDICTIBILIDAD        │
└─────────────────────────────────────────────────────────────────────┘

🎯 ÚSALO CUANDO NECESITES:
   ✅ Generar contenido para integración automática
   ✅ Formato exacto (JSON, CSV, HTML, Markdown)
   ✅ Longitud controlada y predecible
   ✅ Estructura que se repita siempre igual
   ✅ Procesar salidas con código (parsing automático)

❌ EVÍTALO CUANDO:
   ❌ Quieras máxima flexibilidad creativa
   ❌ Busques respuestas breves y directas
   ❌ Necesites ahorrar tokens al máximo
""")
````

### 3️⃣ **Few-shot (ejemplos en prompt)** 📚

**📝 Explicación (Personal):** Es ayudar a IA con ejemplos de salida para guiarla a lo que quieres.

````python
import os

from openai import OpenAI

api_key = os.getenv("Key")
base_url = os.getenv("base_url")
client = OpenAI(base_url=base_url, api_key=api_key)

print("="*80)
print("📚 TÉCNICA: FEW-SHOT (Ejemplos en el Prompt)")
print("="*80)

print("\n🔬 EJEMPLO PRÁCTICO: Clasificación de Sentimientos")
print("-" * 80)

prompt_sin_ejemplos = """Clasifica el siguiente texto según su sentimiento (Positivo, Negativo o Neutral):
"La comida estaba aceptable pero el servicio fue muy lento"
"""

prompt_con_ejemplos = """Clasifica textos según su sentimiento. Aquí están los EJEMPLOS:

EJEMPLO 1:
Texto: "¡Adoré el café! El ambiente es perfecto y el barista es increíble"
Clasificación: Positivo

EJEMPLO 2:
Texto: "El producto llegó dañado y tardaron días en responder"
Clasificación: Negativo

EJEMPLO 3:
Texto: "El libro tiene 200 páginas"
Clasificación: Neutral

Ahora clasifica este texto:
Texto: "La comida estaba aceptable pero el servicio fue muy lento"
Clasificación:
"""

comparaciones = [
    ("ZERO-SHOT (SIN ejemplos)", prompt_sin_ejemplos),
    ("FEW-SHOT (CON 3 ejemplos)", prompt_con_ejemplos),
]

respuestas_fs = {}

for titulo, prompt in comparaciones:
    print(f"\n{'='*80}")
    print(f"🎯 {titulo}")
    print('='*80)
    print(f"\n📋 PROMPT:\n{prompt[:180]}...")
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=200,
        temperature=0.3
    )
    
    respuesta = response.choices[0].message.content
    respuestas_fs[titulo] = respuesta
    
    print(f"\n💬 RESPUESTA:")
    print("-" * 80)
    print(respuesta)

print(f"\n{'='*80}")
print("🔍 EJEMPLO ADICIONAL: Extracción de Datos (Nombres y Edades)")
print('='*80)

prompt_extraction = """Extrae nombre y edad. Formato: Nombre | Edad

EJEMPLO 1:
Texto: "Mi amigo Carlos tiene 28 años y es ingeniero"
Salida: Carlos | 28

EJEMPLO 2:
Texto: "María cumple 35 este mes, es doctora"
Salida: María | 35

Ahora tu turno:
Texto: "Juan acaba de cumplir 42 años y trabaja en tecnología"
Salida:
"""

print(f"\n📋 PROMPT:\n{prompt_extraction[:150]}...")
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt_extraction}],
    max_tokens=50,
    temperature=0.3
)

respuesta_extraction = response.choices[0].message.content
print(f"\n💬 RESPUESTA:\n{respuesta_extraction}")

print("\n" + "="*80)
print("📊 ANÁLISIS COMPARATIVO VISUAL:")
print("="*80)

zero_resp = respuestas_fs.get("ZERO-SHOT (SIN ejemplos)", "")
few_resp = respuestas_fs.get("FEW-SHOT (CON 3 ejemplos)", "")

print(f"""
┌─────────────────────────────────────────────────────────────────────┐
│ COMPARACIÓN: ZERO-SHOT vs FEW-SHOT                                  │
└─────────────────────────────────────────────────────────────────────┘

1️⃣ ZERO-SHOT (SIN ejemplos):
   ├─ Patrón claro:     ❌ NO (modelo debe asumir)
   ├─ Consistencia:     ❌ BAJA (puede variar mucho)
   ├─ Formato:          ❌ IMPREDECIBLE
   ├─ Ambigüedad:       ❌ ALTA (¿qué quería exactamente?)
   ├─ Respuesta:        "{zero_resp.strip()}"
   └─ Riesgo:           ❌ ALTO de no ser lo esperado

2️⃣ FEW-SHOT (CON 3 ejemplos):
   ├─ Patrón claro:     ✅ SÍ (modelo ve exactamente)
   ├─ Consistencia:     ✅ ALTÍSIMA (muy predecible)
   ├─ Formato:          ✅ GARANTIZADO
   ├─ Ambigüedad:       ✅ ELIMINADA (patrón cristalino)
   ├─ Respuesta:        "{few_resp.strip()}"
   └─ Riesgo:           ✅ MÍNIMO (sigue el patrón)

3️⃣ EXTRACCIÓN CON FEW-SHOT:
   ├─ Resultado:        "{respuesta_extraction.strip()}"
   ├─ Formato:          ✅ PERFECTO (Juan | 42)
   ├─ Procesable:       ✅ SÍ (máquinas pueden parsearlo)
   └─ Confianza:        ✅ 100%

┌─────────────────────────────────────────────────────────────────────┐
│ DIFERENCIA CRUCIAL: FEW-SHOT convierte tareas ambiguas en         │
│ tareas DETERMINÍSTICAS (siempre mismo formato)                     │
└─────────────────────────────────────────────────────────────────────┘

🎯 ÚSALO CUANDO NECESITES:
   ✅ Formato CONSISTENTE (2-3 ejemplos suficientes)
   ✅ Evitar AMBIGÜEDAD en interpretación
   ✅ Clasificación de categorías
   ✅ Extracción de datos estructurados
   ✅ Integración automática downstream
   ✅ Traducción con estilo específico

❌ NO NECESITAS CUANDO:
   ❌ Tarea es MUY OBVIA (traducir, resumir)
   ❌ Quieres máxima FLUIDEZ creativa
   ❌ Buscas AHORRAR TOKENS
   ❌ Respuestas abiertas/generativas
""")
````

### 5️⃣ **Cues / priming del output** 🎯
